# Loss Function 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wrn
wrn.filterwarnings("ignore")
```

```python
p = np.arange(0, 1, 0.01)
```

```python
# Cost function if y_actual = 1

plt.plot(p, -1 * np.log(p))
plt.show()
```

```python
# Cost function if y_actual = 0

plt.plot(p, -1 * (np.log(1 - p)))
plt.show()
```

```python
y_actual = [0, 1, 1, 0, 1]
y_pred = [0.3, 0.2, 0.7, 0.6, 0.7]
```

### Formula Log Loss 

### Create it from scratch 

```python
N = 5
loss_func = 0
for i in range(N):
    
    loss_func += (-1/N) * (y_actual[i] * np.log(y_pred[i]) + (1-y_actual[i]) * np.log(1-y_pred[i]))
    
print(loss_func)
```

### Compare it from Scikit-learn 

```python
from sklearn.metrics import log_loss
```

```python
log_loss(y_actual, y_pred)
```
